```python
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
data={'surname':['Smith', 'Johnson', 'Lee', 'Kim', 'Garcia', 'Wang', 'Müller', 'Lopez', 'ram'],
     'country':['England', 'England', 'China', 'China', 'Spain', 'China', 'Germany', 'Spain','India']}

df=pd.DataFrame(data)
df
```

```python
ohe=OneHotEncoder(sparse=False)
surname_encoded=ohe.fit_transform(df['surname'].values.reshape(-1,1))
surname_encoded
```

```python
df_encoded=pd.concat([df,pd.DataFrame(surname_encoded)],axis=1)
df_encoded
```

```python
X=df_encoded.iloc[:,2:]
Y=df_encoded['country']
X
```

```python
le=LabelEncoder()
Y=le.fit_transform(Y)
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(150,100,50,10), max_iter=1000, activation='relu', solver='adam')
mlp.fit(X_train,Y_train)
```

```python
unseen_surnames = ['Smith', 'Garcia', 'Wang','Lopez','ram']
unseen_surnames_encoded=ohe.transform(pd.DataFrame(unseen_surnames).values)
country_idx_pred=mlp.predict(unseen_surnames_encoded)
country_pred=le.inverse_transform(country_idx_pred)

for surnames, country in zip(unseen_surnames,country_pred):
    print(surnames, "->", country)
```
